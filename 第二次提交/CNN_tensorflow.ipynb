```python
import tensorflow as tf
import numpy as np

print("TensorFlow 版本:", tf.__version__)

# 加载 MNIST 数据
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# 归一化
x_train, x_test = x_train / 255.0, x_test / 255.0

# 重塑数据为 (batch, height, width, channels)
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

print(f"训练数据形状: {x_train.shape}")
print(f"测试数据形状: {x_test.shape}")
```

```python
# 超参数
learning_rate = 1e-4
keep_prob_rate = 0.7
max_epochs = 5
BATCH_SIZE = 100

print("构建 CNN 模型...")

# 使用 Keras Sequential API 构建 CNN
model = tf.keras.models.Sequential([
    # 卷积层 1: 7x7 卷积核, 32 个滤波器
    tf.keras.layers.Conv2D(32, kernel_size=7, strides=1, padding='same',
                          activation='relu', input_shape=(28, 28, 1)),
    tf.keras.layers.MaxPooling2D(pool_size=2, strides=2, padding='same'),

    # 卷积层 2: 5x5 卷积核, 64 个滤波器
    tf.keras.layers.Conv2D(64, kernel_size=5, strides=1, padding='same',
                          activation='relu'),
    tf.keras.layers.MaxPooling2D(pool_size=2, strides=2, padding='same'),

    # 展平层
    tf.keras.layers.Flatten(),

    # 全连接层 1: 1024 个神经元
    tf.keras.layers.Dense(1024, activation='relu'),
    tf.keras.layers.Dropout(keep_prob_rate),

    # 全连接层 2 (输出层): 10 个神经元
    tf.keras.layers.Dense(10, activation='softmax')
])

# 显示模型结构
model.summary()
```

```python
# 编译模型
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# 训练模型
print("开始训练...")
print("=" * 60)

history = model.fit(x_train, y_train,
                    batch_size=BATCH_SIZE,
                    epochs=max_epochs,
                    validation_data=(x_test[:1000], y_test[:1000]),
                    verbose=1)
```

```python
# 最终评估
print("=" * 60)
print("训练完成！")

test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
print(f"最终测试集准确率: {test_acc:.4f}")

if test_acc >= 0.96:
    print("[SUCCESS] 达到目标精度 96% 以上！")
else:
    print("[WARNING] 未达到目标精度 96%，可能需要增加训练轮数")

print("=" * 60)
```
